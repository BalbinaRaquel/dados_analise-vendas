### 4.1. Agrupamento de Dados

```python
tabela_vendas.head()
```

```python
# Faturamento por loja
tabela_vendas.groupby("loja")["preco"].sum().to_frame()
```

```python
# Ticket médio por loja
tabela_vendas.groupby("loja")["preco"].mean().to_frame()
```

```python
# faturamento por loja e cidade
tabela_vendas.groupby(["loja", "cidade"])["preco"].sum().to_frame()
```

```python
# faturamento por loja e forma de pagamento
tabela_vendas.groupby(["loja", "forma_pagamento"])["preco"].sum().to_frame()
```
